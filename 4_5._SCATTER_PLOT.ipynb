```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/merc.csv')
df.head()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
#sns.scatterplot(data=df, x='mileage', y='price')
sns.regplot(data=df, x='year', y='price')   # osuvchi karilatsiya
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='mileage', y='price', hue='transmission', s=100, style='fuelType') # hue->ranglar bn ajratib beradi s=>nuqtalar hajmi
#sns.regplot(data=df, x='year', y='price')   # osuvchi karilatsiya       # style qoshimcha parametr
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price']) #
plt.show()
```
