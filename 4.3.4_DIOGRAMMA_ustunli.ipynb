```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates=[0], usecols=[0,1])
df.head()
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index] # dfda month degan ustun yaratib unga df.indexdan har bir oyni sugirib yukla deyabmiz
df.head()
```

```python
df['year'] = [i.year for i in df.index]
df.head()
```

```python
df2 = df.groupby(['month','year']).max()
df2.head()
```

```python
df2.unstack(level=0) # unstack orqali bitta ustunni index holatga aylantiramiz va uni level=0 yoki 1 qilish orqali amalga oshiramiz
```

```python
plt.figure(figsize=(16, 6))
sns.heatmap(data=df2.unstack(level=0), cmap= "Blues", linewidths=1, square=True)
xticks_labels=['jan','feb', 'mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
plt.xticks(np.arange(12)+.5, labels=xticks_labels ) # oylarni alohida yozib chiqdik 
plt.show()
```
