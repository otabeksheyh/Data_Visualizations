```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize=(16,8))

plt.plot(x, y, linewidth=2, linestyle="--", color="r", marker='o')
plt.title("sin(x)  funksiya")
plt.xlabel('pi')
plt.ylabel("sin(x)")
plt.grid()
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin =np.sin(x)
cos = np.cos(x)
```

```python
plt.figure(figsize=(16,6))
plt.plot( x, sin, linewidth=1.5, linestyle="--", color='r', alpha=0.7,marker='o')
plt.plot(x, cos, linewidth=1.5, linestyle=":", color='b', alpha=1, marker='h')
plt.title('sin(x) va cos(x) funksiya grafigi')
plt.xlabel("pi")
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.show()
```

```python
df= pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv', index_col=0, parse_dates=[0],usecols=[0,1])
df.head()
```

```python
type(df.index)
```

```python
df['month'] = [i.month for i in df.index]
df.head()
```

```python
df['year'] = [i.year for i in df.index]
```

```python
df2 = df.groupby(['month', 'year']).max()
df2
```

```python
df2.unstack(level=0)
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(data=df2.unstack(level=0), cmap='Blues', linewidths=1, square=True)
xticks_labels = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

plt.xticks(np.arange(12)+ .5, labels=xticks_labels)
plt.show()
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/merc.csv')
df.head()
```

```python
df = df[df.year>=2010]
df =df[df.transmission!='Other']
df=df[df.fuelType!='Other']
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(16,6))
sns.histplot(data=df, x='price')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.displot(data=df, x='price',hue='transmission',multiple='stack', height=5, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.displot(data=df, x='price',col='transmission', height=15, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price',kde=True, height=15, aspect=2)
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.kdeplot(data=df, x='price', hue='transmission', shade=True)
plt.show()
```

```python
sns.histplot(data=df, x='price')       
plt.show()
```

```python
sns.scatterplot(data=df, x='price', y='year', s=100)
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,10)) 
sns.histplot(ax=ax[0], data=df, x ='price')

sns.scatterplot(ax=ax[1],data=df, x ='price', y='year', s=100)


plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,10)) 

fig.suptitle('GRAFIK 2')

sns.histplot(ax=ax[0, 0], data=df, x='price')
sns.histplot(ax=ax[0, 1],  data=df, x='mileage')

sns.scatterplot(ax=ax[1,0], data=df, x=df['year'], y=df['price'], hue='fuelType', s=50)
sns.scatterplot(ax=ax[1, 1], data=df, x=df['year'], y=df['mileage'], hue='fuelType', s=50)

ax[0,0].set_title("price distribution")
ax[0,1].set_title("mileage distributon")
ax[1,0].set_title("price vs year distribution")
ax[1,1].set_title("mileage vs year distribution")


plt.show()
```
