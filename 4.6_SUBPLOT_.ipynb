import numpy as np

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/merc.csv")
df = df[df.year>=2010]
df =df[df.transmission!='other']
df =df[df.fuelType!='other']
df.head()
```

```python
x=np.arange(0, 4*np.pi, 0.1)
sin=np.sin(x)
cos=np.cos(x)

fig1 = plt.figure(figsize=(10,6))

plt.subplot(2,1,1)

plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color = 'y',
         alpha=0.7,
         marker='h')
plt.title("sin(x)funksiya grafigi")
plt.grid(color='tomato', alpha=0.5, linestyle='--')


plt.subplot(2,1,2)
plt.plot(x, cos, 
         linewidth=2, 
         linestyle=':', 
         color='b', 
         alpha=0.7, 
         marker='v')

plt.title("cos(x) funksiya grafigi")
plt.grid(color='tomato', alpha=0.5, linestyle='--')

plt.show()
```

```python
sns.histplot(data=df, x='price')
plt.show()
```

```python
sns.scatterplot(data=df, x='price', y='year', s=100)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10,10))
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, x='price', y='year', s=100)
ax[0].set_title("price distribution")
ax[1].set_title("year vs price distribution")
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(15,10))


fig.suptitle("GRAFIK 2")
sns.histplot(ax=ax[0,0], data=df, x='price')
sns.histplot(ax=ax[0,1], data=df, x='mileage')
sns.scatterplot(ax=ax[1,0], data=df, x=df['year'], y=df['price'], hue='fuelType', s=50)
sns.scatterplot(ax=ax[1,1], data=df,x=df['year'], y=df['mileage'],hue='fuelType', s=50)
ax[0,0].set_title("price distribution")
ax[0,1].set_title("mileage distributon")
ax[1,0].set_title("price vs year distribution")
ax[1,1].set_title("mileage vs year distribution")


plt.show()
```
