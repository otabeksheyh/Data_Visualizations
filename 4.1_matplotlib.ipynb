```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x,y, linewidth=2, linestyle="--", color="r", marker='>')
plt.title("sin(x) funksiyasi")
plt.xlabel("pi")
plt.ylabel("sin(x)")
plt.grid(color="b")

plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)


plt.figure(figsize=(12, 6))
plt.plot(x, sin, 
         linewidth=1.5, 
         linestyle="-.", 
         color='#8E44AD',
         alpha=0.7, 
         marker="h")

plt.plot(x, cos, 
         linewidth=1.5, 
         linestyle=':', 
         color='c', 
         alpha=0.7, 
         marker="v")


plt.title('sin(x) va cos(x) funksiya grafigi')
plt.xlabel("pi")
plt.legend(['sin(x)','cos(x)'], loc=1, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--')
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# Grafik nomi:
plt.title('sin(x) va cos(x) funksiya grafigi')
# x-o'q nomi:
plt.xlabel("pi")
# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.show()

```
